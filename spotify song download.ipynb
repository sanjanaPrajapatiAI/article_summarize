```python
import os
def download_spotify_playlist(playlist_url,output_floder):
  os.makedirs(output_floder,exist_ok=True)
  command=f"spotdl{playlist_url} -o {output_floder}"
  os.system(command)

playlist_url=input("Enter the spotify playlist url:")
output_floder=input("Enter the output floder:")
download_spotify_playlist(playlist_url,output_floder)
```
