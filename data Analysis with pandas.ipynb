```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
path='/content/sample_data/california_housing_train.csv'
df=pd.read_csv(path)

ax = sns.countplot(x='latitude', data=df)
ax=sns.countplot(x='longitude',data=df)
plt.show()
df['latitude']=df['latitude'].astype('int64')
column_data=df['latitude']
sns.distplot(column_data)
plt.show()
```
